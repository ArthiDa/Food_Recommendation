## Text Vectoraization -> Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=1500,stop_words='english')
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
meals['tags'].apply(stem)
```

```python
meals['tags'] = meals['tags'].apply(stem)
```

```python
meals.iloc[0].tags
```

```python
cv.fit_transform(meals['tags']).toarray()
```

```python
vectors = cv.fit_transform(meals['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

## Find the similar meals using cosine distance
